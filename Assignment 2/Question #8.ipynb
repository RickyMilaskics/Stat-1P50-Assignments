```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('Data/student.csv')
data = data[['studytime', 'grade']]



index = data['studytime'] < 3
listOfGradesLessThan3StudyTime = list(data.loc[index, 'grade'])

index = data['studytime'] >= 3
listofGradesMoreThan3StudyTime = list(data.loc[index, 'grade'])

data = [listOfGradesLessThan3StudyTime,listofGradesMoreThan3StudyTime]

plt.title('1-2 hour studytime vs 3-4 hour studytime to grades')
plt.boxplot(data)

plt.show()
```
