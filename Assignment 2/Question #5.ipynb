```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Data/student.csv')

data = data[['studytime', 'grade']]

data2 = data.groupby('studytime').mean().reset_index()

print(data2)

#convert the columns into their own lists
data2x = data2['studytime'].tolist()
data2y = data2['grade'].tolist()

#make the line graph and display it
plt.plot(data2x,data2y)
plt.show()
```
