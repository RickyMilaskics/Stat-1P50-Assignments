```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('Data/student.csv')
data = data[['studytime', 'grade']]



index = data['studytime'] < 3
listOfGradesLessThan3StudyTime = list(data.loc[index, 'grade'])

index = data['studytime'] >= 3
listofGradesMoreThan3StudyTime = list(data.loc[index, 'grade'])


plt.title('grades of students with studytime less than 3')
plt.boxplot(listOfGradesLessThan3StudyTime)
plt.show()

plt.title('grades of students with studytime more than 3')
plt.boxplot(listofGradesMoreThan3StudyTime)
plt.show()
```
