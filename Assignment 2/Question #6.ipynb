```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('Data/crime.csv')

#convert the comlumns into lists to show on a scatter plot
datax = data['PctPopUnderPov'].tolist()
datay = data['ViolentCrimesPerPop'].tolist()
plt.scatter(datax,datay)
plt.show()

#convert the columns into an array
datay = np.array(data['ViolentCrimesPerPop']).reshape(-1,1)
datax = np.array(data['PctPopUnderPov']).reshape(-1,1)

#scale the values
scaler = StandardScaler()
datax = scaler.fit_transform(datax)
datay = scaler.fit_transform(datay)

#show the new scatter plot with the new values
plt.scatter(datax,datay)
plt.show()
```
