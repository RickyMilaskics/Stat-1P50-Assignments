```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Data/student.csv')

data = data[['internet', 'grade']]

data2 = data.groupby('internet').mean().reset_index()

print(data2)

#convert the columns into their own lists
data2x = ['students with no internet', 'students with internet']
data2y = data2['grade'].tolist()

#make the bar graph and display it
plt.bar(x=data2x,height=data2y)
plt.show()
```
